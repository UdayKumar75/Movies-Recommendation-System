```python
movies.dropna(axis = 0, inplace = True)
```

```python
import ast

def convert(obj):
    list = []
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list
```

```python
movies.genres = movies.genres.apply(convert)
```

```python
movies.head()
```

```python
movies.keywords = movies.keywords.apply(convert)
movies.production_companies = movies.production_companies.apply(convert)
```

```python
movies.head()
```

```python
def convert2(obj):
    list = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 5:
            list.append(i['name'])
            count+=1
        else:
            break
    return list
```

```python
movies.cast = movies.cast.apply(convert2)
```

```python
movies.head()
```

```python
movies.cast[0]
```

```python
def convert3(obj):
    list = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            list.append(i['name'])
            break
    return list
```

```python
movies.crew = movies.crew.apply(convert3)
```

```python
movies.head()
```

```python
movies.overview = movies.overview.apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies.genres = movies.genres.apply(lambda x:[i.replace(" ","") for i in x])
movies.keywords = movies.keywords.apply(lambda x:[i.replace(" ","") for i in x])
movies.production_companies = movies.production_companies.apply(lambda x:[i.replace(" ","") for i in x])
movies.cast = movies.cast.apply(lambda x:[i.replace(" ","") for i in x])
movies.crew = movies.crew.apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies.overview + movies.genres + movies.keywords + movies.cast + movies.crew + movies.production_companies
```

```python
movies = movies[['movie_id', 'title', 'tags']]
movies.head()
```

```python
movies.tags = movies.tags.apply(lambda x: " ".join(x))
```

```python
movies.tags[0]
```

```python
movies.tags = movies.tags.apply(lambda x: x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies.tags = movies.tags.apply(stem)
```

```python
movies.tags[0]
```

```python
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(movies.tags).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vectors)
```

```python
sim
```

```python
def recommend(movie):
    movie_index = movies[movies['title'] == movie].index[0]
    dist = sim[movie_index]
    movies_list = sorted(list(enumerate(dist)), reverse = True, key = lambda x:x[1])[1:6]

    for i in movies_list:
        print(movies.iloc[i[0]].title)
```
